```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
# 聚类数据生成器
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
data
```

```python
sns.scatterplot(x=data[0][:, 0], y=data[0][:, 1], hue=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X=data[0])
```

```python
kmeans.cluster_centers_
```

###### True Label. Normally, K Means is going to give you labels

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(10,6))

ax1.set_title("K Means")
ax1.scatter(x=data[0][:,0], y=data[0][:,1], c=kmeans.labels_, cmap="rainbow", edgecolors="black")

ax2.set_title("Original")
ax2.scatter(x=data[0][:,0], y=data[0][:,1], c=data[1], cmap="rainbow", edgecolors="black")
```

```python
kmeans = KMeans(n_clusters=8)
```

```python
kmeans.fit(X=data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```
